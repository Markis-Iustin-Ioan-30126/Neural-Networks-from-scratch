# Activation functions visualization

```python
import numpy as np
from activations import *
import matplotlib.pyplot as plt
```

### Sigmoid 

```python
x = np.arange(-8, 8, 0.01)

plt.plot(x, sigmoid(x))
plt.plot(x, sigmoid_gradient(x))
plt.title("Sigmoid activation function")
plt.xlabel("x")
plt.grid()
plt.legend((("f(x)"), ("f'(x)")))
```

### Hyperbolic tangent

```python
x = np.arange(-5, 5, 0.01)

plt.plot(x, tanh(x))
plt.plot(x, tanh_gradient(x))
plt.grid()
plt.title("Hyperbolic tangent activation function")
plt.xlabel("x")
plt.legend((("f(x)"), ("f'(x)")))
```

### ReLU activations
---
1) ReLU
2) Parametric ReLU
3) Exponential ReLU

```python
x = np.array(np.arange(-3, 3, 0.01), ndmin=2).T

plt.figure()
plt.plot(x, relu(x))
plt.plot(x, relu_gradient(x))
plt.grid()
plt.title("Rectified linear unit activation function")
plt.xlabel("x")
plt.legend((("f(x)"), ("f'(x)")))

plt.figure()
plt.plot(x, param_relu(x))
plt.plot(x, param_relu_gradient(x))
plt.grid()
plt.title("Parametric rectified linear unit activation function")
plt.xlabel("x")
plt.legend((("f(x)"), ("f'(x)")))

plt.figure()
plt.plot(x, elu(x, 0.5))
plt.plot(x, elu_gradient(x, 0.5))
plt.grid()
plt.title("Exponential rectified linear unit activation function")
plt.xlabel("x")
plt.legend((("f(x)"), ("f'(x)")))
```
